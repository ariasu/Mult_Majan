```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

df = pd.read_excel('majanTest_crean.xlsx',index_col=0)

df2 = df.reindex(columns=["記録対戦数","和了巡数","平均和了","平均放銃","和了率","放銃率","ツモ率","ダマ率","流局率","流局聴牌率","副露率","立直率","飛び率","平均順位"])
df2
```

```python
a = -1
x = df2.iloc[:,:a]
y = df2.iloc[:,a]

model = LinearRegression()
model.fit(x,y)

np.set_printoptions(precision=4,suppress=True)
model.coef_
```

```python
scaler = StandardScaler()
scaler.fit(x)
x2 = scaler.transform(x)
x2
```

```python
model2 = LinearRegression()
model2.fit(x2,y)
model2.coef_
#和了率が一番高い
#２番目は平均放銃
#３番目は平均和了
```
