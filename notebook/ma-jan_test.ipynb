## 何が和了率に一番影響をおよぼしているのか？

```python
X = df3.iloc[:,:-3]
Y = df3.iloc[:,-3]
```

```python
scaler = StandardScaler()
scaler.fit(X)
X2 = scaler.transform(X)
X2
```

```python
model3 = LinearRegression()
model3.fit(X2,Y)
model3.coef_
```
