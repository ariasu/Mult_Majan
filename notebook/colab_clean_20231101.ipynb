```python
from google.colab import drive
drive.mount('/content/drive')

%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from pathlib import Path
import re

path_base = Path("/content/drive/MyDrive/Colab Notebooks/麻雀")
filepath_xlsx = path_base.joinpath("df_rework_test.xlsx")

List_columName = ['記録対戦数', '和了巡数', '平均和了', '平均放銃', '平均順位',
                  '和了率', '放銃率', 'ツモ率', 'ダマ率', '流局率',
                  '流局聴牌率', '副露率', '立直率', '飛び率', '記録点数']

df = pd.read_excel(filepath_xlsx,index_col=0)
df = df.iloc[:300,:18]
df = df.drop(["記録段位", "安定段位", "点数期待"], axis=1)
df= df.reindex(columns=List_columName)
display(df.shape)
display(df.dtypes)
display(df.head(3))
```

```python
x = len(df.columns)
y = len(df.index)
data_void = pd.DataFrame(np.arange(x*y).reshape(y,x), index=df.index, columns=List_columName)

for h in range(5):
    a = df.iloc[:,h].astype(float)
    for i in range(len(df)):
        data_void.iloc[i,h] = a[i]

for d in range(5,14):
    for per in range(0,300):
        s = df.iloc[per,d]
        b = re.sub('%$', '', s)
        data_void.iloc[per,d] = float(b)

for r in range(5,14):
    for c in range(0,300):
        data_void.iloc[c,r] = data_void.iloc[c,r] * 0.01

data_void = data_void.drop("記録点数", axis=1)
data_void = data_void.drop_duplicates() # 重複した行を削除

data_clean = data_void[data_void['記録対戦数']>99]
filepath_output = path_base.joinpath("df_rework_test_clean.xlsx")
data_clean.to_excel(filepath_output)
pd.read_excel(filepath_output)
```
