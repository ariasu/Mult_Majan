```python
from selenium import webdriver
import pandas as pd
import numpy as np
browser = webdriver.Chrome('./chromedriver')
df = pd.read_excel("df_sum_rework.xlsx",index_col=0)

browser.get('https://saki.sapk.ch/ranking/career1')
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(browser,1.5)
wait2 = WebDriverWait(browser,0)

url = 'https://saki.sapk.ch/ranking/career1'
browser.get(url)

#ランキング名を拾う
rank_name = '/html/body/div/div/div[2]/div[2]/a[2]'
rank_name2 = wait. until(EC.visibility_of_element_located((By.XPATH, rank_name)))
browser.execute_script("arguments[0].click();", rank_name2)

#300戦を拾う
b1 = browser.find_element_by_xpath('/html/body/div/div/div[2]/fieldset/div/label[1]/span[1]/input')
browser.execute_script("arguments[0].click();", b1)

for youser_num in range(1,101):
    #ユーザー名をクリック
    youser_name = '/html/body/div/div/div[2]/div[4]/div[1]/div/table/tbody/tr[{}]/td[2]/a'.format(youser_num)
    youser_name2 = wait. until(EC.visibility_of_element_located((By.XPATH, youser_name)))
    browser.execute_script("arguments[0].click();", youser_name2)

    next_result_num = 0

    for simple_num in (1,2,3,5):
        #基本、立直、ほか,を拾う
        simple = '/html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/a[{}]'.format(simple_num)
        simple2 = wait2. until(EC.visibility_of_element_located((By.XPATH, simple)))
        browser.execute_script("arguments[0].click();", simple2)

        #項目数を調べる

        koumoku = 'html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[3]/div[1]/h6'
        koumoku2 = wait2. until(EC.visibility_of_element_located((By.XPATH, koumoku)))
        koumoku_un_inf = 1
        koumoku3 = []

        try:
            while koumoku2 and koumoku_un_inf <= 20:
                koumoku3.append(koumoku2.text)

                koumoku_un_inf += 1
                koumoku = 'html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[3]/div[{}]/h6'.format(koumoku_un_inf)
                koumoku2 = wait2. until(EC.visibility_of_element_located((By.XPATH, koumoku)))

        except:
            print(simple2.text,"の項目数は",len(koumoku3),"です")

        for result_num in range(1,len(koumoku3)+1):
            try:
                #リザルトを拾う
                result = '/html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[3]/div[{}]/p'.format(result_num)
                result2 = wait2. until(EC.visibility_of_element_located((By.XPATH, result)))
            except:
                result2 = "0"

            #dfに入れる
            df.iloc[youser_num-1,next_result_num+result_num-1] = result2.text
            print("ここに入れました",youser_num-1,next_result_num+result_num-1)
            print("その中身は",df.iloc[youser_num-1,next_result_num+result_num-1])
            print("---")

        next_result_num = next_result_num + len(koumoku3)
        print("next_result_numは",next_result_num,"です")
        
    url = 'https://saki.sapk.ch/ranking/career1'
    browser.get(url)
    
df
```
