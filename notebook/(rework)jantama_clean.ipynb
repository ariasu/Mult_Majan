```python
import pandas as pd
import numpy as np

df = pd.read_excel("df_rework_test.xlsx",index_col=0)
```

```python
df2 = df.iloc[0:300,0:18]
df2
```

```python
df3 = df2.drop("記録段位", axis=1)
df4 = df3.drop("安定段位", axis=1)
df5 = df4.drop("点数期待", axis=1)
df5
```

```python
data = df5.reindex(columns=['記録対戦数', '和了巡数', '平均和了', '平均放銃', '平均順位', '和了率', '放銃率', 'ツモ率', 'ダマ率', '流局率', '流局聴牌率', '副露率',
       '立直率', '飛び率', '記録点数'])
data
```

```python
x = len(data.columns)
y = len(data.index)
data_void = pd.DataFrame(np.arange(x*y).reshape(y,x),index=data.index,columns=['記録対戦数', '和了巡数', '平均和了', '平均放銃', '平均順位', '和了率', '放銃率', 'ツモ率', 'ダマ率', '流局率', '流局聴牌率', '副露率','立直率', '飛び率','記録点数'])

data_void
```

```python
for h in range(0,5):
    a = data.iloc[:,h].astype(float)

    for i in range(0,300):
        data_void.iloc[i,h] = a[i]
        
data_void
```

```python
import re

for d in range(5,14):
    
    for per in range(0,300):
        s = data.iloc[per,d]

        b = re.sub('%$', '', s)
        data_void.iloc[per,d] = float(b)
    
data_void
```

```python
for r in range(5,14):
    for c in range(0,300):
        data_void.iloc[c,r] = data_void.iloc[c,r] * 0.01
data_void
```

```python
data_void2 = data_void.drop("記録点数", axis=1)
data_void2
```
