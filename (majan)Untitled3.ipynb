```python
import numpy as np
import pandas as pd
import re

df = pd.read_excel('majanTest(20220129).xlsx',index_col=0)
```

```python
df2 = df.drop("安定段位", axis=0)
df3 = df2.drop("記録段位", axis=0)
df4 = df3.drop("記録点数", axis=0)
df5 = df4.drop("点数期待", axis=0)
df5
```

```python
#"平均和了","平均放銃","平均順位","和了率","放銃率","ツモ率","ダマ率","流局率","流局聴牌率","副露率","立直率"
data = df5.reindex(index=["記録対戦数","和了巡数","平均和了","平均放銃","平均順位","和了率","放銃率","ツモ率","ダマ率","流局率","流局聴牌率","副露率","立直率","飛び率"])
data
```

```python
data.columns
```

```python
#data.index
#["記録対戦数","和了巡数","平均和了","平均放銃","平均順位","和了率","放銃率","ツモ率","ダマ率","流局率","流局聴牌率","副露率","立直率","飛び率"]
y = len(data.columns)
x = len(data.index)
data_void = pd.DataFrame(np.arange(x*y).reshape(y,x),index=data.columns,columns=["記録対戦数","和了巡数","平均和了","平均放銃","平均順位","和了率","放銃率","ツモ率","ダマ率","流局率","流局聴牌率","副露率","立直率","飛び率"])

data_void
```

```python
for h in range(0,5):
    a = data.iloc[h,:].astype(float)

    for i in range(0,100):
        data_void.iloc[i,h] = a[i]
        
data_void
```

```python
for d in range(5,14):
    
    for per in range(0,100):
        s = data.iloc[d,per]

        b = re.sub('%$', '', s)
        data_void.iloc[per,d] = float(b)
    
data_void
```

```python
for r in range(5,14):
    for c in range(0,100):
        data_void.iloc[c,r] = data_void.iloc[c,r] * 0.01
data_void
```

```python
data_void.to_excel("majanTest_crean.xlsx")
```
