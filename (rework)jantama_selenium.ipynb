## 項目を入れる（その数を使って空のデータベース作る）

```python
browser.get('https://saki.sapk.ch/ranking/career1')
title = []
stack = 0

#項目を入れる
wow = browser.find_element_by_xpath('/html/body/div/div/div[2]/div[4]/div[1]/div/table/tbody/tr[1]/td[2]/a')
wow.click()

time.sleep(1)

for a in range(1,8):
    last = browser.find_element_by_xpath('/html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/a[{}]'.format(a))
    browser.execute_script("arguments[0].click();", last)

    name = []
    if a != 4 :
        try:
            for b in range(1,19):
                text2 = browser.find_element_by_xpath('/html/body/div[1]/div/div[2]/div[1]/div[1]/div/div[1]/div[3]/div[{}]/h6'.format(b))
                name.append(text2)
        except:
            b = b - 1
        
        maax = b + 1
        maaax = b
        
        for c in range(1,maax):
            title.append(0)
    
        try:
            for d in range(0,maax):
                title[stack+d] = name[d].text
        except:
            stack = stack + d
title 
```

## プレイヤー名を集めよう！

```python
#プレイヤー名を入れる
browser.get('https://saki.sapk.ch/ranking/career1')
player = []

for a in range(1,16):
    millon = browser.find_element_by_xpath('/html/body/div/div/div[2]/div[2]/a[{}]'.format(a))
    browser.execute_script("arguments[0].click();", millon)
    
    if a != 1 and a != 4 and a != 11 and a != 12 and a != 13:
        for b in range(1,4):
            millon2 = browser.find_element_by_xpath('/html/body/div/div/div[2]/fieldset/div/label[{}]/span[1]/input'.format(b))
            browser.execute_script("arguments[0].click();",millon2)
            time.sleep(0.5)
    
            for c in range(1,101):
                wow = browser.find_element_by_xpath('/html/body/div/div/div[2]/div[4]/div[1]/div/table/tbody/tr[{}]/td[2]/a'.format(c))
                player.append(wow.text)

player
```

## 2次元配列を作る

```python
x = len(title)
y = len(player)
df_sum = pd.DataFrame(np.arange(x*y).reshape(y,x),index=player,columns=title)

df_sum.to_excel("df_sum_rework.xlsx")
```
